```python
df = df.dropna(subset=['name', 'description'])
```

```python
df.isnull().sum()
```

```python
df['Promotion'] = df['Promotion'].str.strip()
```

```python
df['Promotion'] = df['Promotion'].map({'Yes' : 1, 'No' : 0})
```

```python
df.info()
```

```python
df['Sales Volume'] = pd.to_numeric(df['Sales Volume'], errors = 'coerce')
```

```python
df.info()
```

```python
df = df.dropna(subset=['Sales Volume', 'Promotion'])
```

```python
promo_group = df.groupby('Promotion')['Sales Volume'].mean()
print(promo_group)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.boxplot(x='Promotion', y = 'Sales Volume', data = df)
plt.title(" vs  ")
plt.xticks([0,1], ['No Promotion', 'Promotion'])
plt.show()
```

```python
from scipy.stats import ttest_ind

promo_yes = df[df['Promotion'] == 1]['Sales Volume']
promo_no = df[df['Promotion'] == 0]['Sales Volume']

t_stat, p_value = ttest_ind(promo_yes, promo_no, equal_var = False)
print(f"T-stat : {t_stat}, P-value : {p_value}")
```
